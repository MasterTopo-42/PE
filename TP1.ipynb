```python
import statistics as st
import numpy as np
```

```python
x = [0,-1,-2,-3]
p = [0.4,0.3,0.2,0.1]

y = [i+2 for i in x]

def prom(x,p):
  return sum(x[i]*p[i] for i in range(len(x)))

def varianza(x,xo,p):
  return sum(p[i]*(x[i]-xo)**2 for i in range(len(x)))

def desv(x,xo,p):
  return varianza(x,xo,p)**(1/2)
```

```python
plt.plot(y,p,'bo',x,p,'ro')
```

```python
x0 = prom(x,p)
y0 = prom(y,p)

Dx = desv(x,x0,p)
Dy = desv(y,y0,p)

print("X\tValor medio:",x0,"\tDesvio estándar:",Dx)
print("Y\tValor medio:",y0,"\tDesvio estándar:",Dy)
```

### Ejercicio 6

Probabilidades: \\
  $0_{ent}=p$ \\
  $1_{ent}=1-p$

cbin.png

```python
p = 0.6
e1 = 0.1
e2 = 0.2
```

```python
#Funcion que determina el bit de entrada (X)
def entrada():
    if random() <= p:
        return 0
    else:
        return 1
```

```python
#Funcion que determina el bit de salida
def salida(X):
    if X == 0:
        if random() <= e1:
            return 1
        else:
            return 0
    else:
        if random() <= e2:
            return 0
        else:
            return 1
```

```python
#Ejemplo (dar al play reiteradamente)
salida(1)
```

```python
N = 1000000
BitsEntrada = [entrada() for i in range(N)]
BitsSalida = [salida(i) for i in BitsEntrada]

#Cantidad de 1's y 0's en la salida
unos = BitsSalida.count(1)
ceros = BitsSalida.count(0)
```

```python
print("Bits de entrada:\t",BitsEntrada,"\n")
print("Bits de salida: \t",BitsSalida)
```

#### Probabilidades simuladas

```python
#Probabilidad de 0 entrada dado 1 salida
def E0_1S(BitsEntrada,BitsSalida):
  c = 0
  for i in range(len(BitsEntrada)):
      if BitsEntrada[i] == 0 and BitsSalida[i] == 1:
          c += 1
  return c/unos

#Probabilidad de 1 entrada dado 1 salida
def E1_1S(BitsEntrada,BitsSalida):
  c = 0
  for i in range(len(BitsEntrada)):
      if BitsEntrada[i] == 1 and BitsSalida[i] == 1:
          c += 1
  return c/unos

#Probabilidad de 0 entrada dado 0 salida
def E0_0S(BitsEntrada,BitsSalida):
  c = 0
  for i in range(len(BitsEntrada)):
      if BitsEntrada[i] == 0 and BitsSalida[i] == 0:
          c += 1
  return c/ceros

#Probabilidad de 1 entrada dado 0 salida
def E1_0S(BitsEntrada,BitsSalida):
  c = 0
  for i in range(len(BitsEntrada)):
      if BitsEntrada[i] == 1 and BitsSalida[i] == 0:
          c += 1
  return c/ceros
```

#### Probabilidades teoricas

```python
z = (p*e1)/((1-p)*(1-e2))
p0e1s = 1/(1+1/z)
p1e1s = 1/(1+z)

w = (p*(1-e1))/(e2*(1-p))
p0e0s = 1/(1+1/w)
p1e0s = 1/(1+w)
```

```python
#Comparaciones
print("Probabilidad\tSimulacion\t\tTeorico\n")
print("P(0_ent/1_sal):\t",E0_1S(BitsEntrada,BitsSalida),"\t",p0e1s,'\n')
print("P(1_ent/1_sal):\t",E1_1S(BitsEntrada,BitsSalida),"\t",p1e1s,'\n')
print("P(0_ent/0_sal):\t",E0_0S(BitsEntrada,BitsSalida),"\t",p0e0s,'\n')
print("P(1_ent/0_sal):\t",E1_0S(BitsEntrada,BitsSalida),"\t",p1e0s,'\n')
```
